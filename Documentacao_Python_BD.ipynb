**2)  Combinar dois ou mais Data Frame.** (método: Concat)

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})

result = pd.concat([df1, df2])
print(result)
```

**3) Permite adicionar um elemento ao final de uma lista existente.** (método: Append)

```python
import pandas as pd

df = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']})

new_row = {'A': 'A4', 'B': 'B4', 'C': 'C4', 'D': 'D4'}
result = df.append(new_row, ignore_index=True)
print(result)
```

# *>Seção 8*: Teoria dos conjuntos, tipos de combinação.

```python
import pandas as pd
```

```python
tabela_1 = pd.DataFrame({
'Nome':['João', 'João', 'Pedro' , 'Caio'],
'Telefone': ['962055302', '963054522', '963057866', '965057814'],
'Carros': ['azul', 'preto', 'verde' , 'amarelo']})

print(tabela_1)
```

```python
tabela_2 = pd.DataFrame({
'Nome':['João', 'Marcelo', 'Thiago' , 'Caio'],
'Irmãos': ['1', '3', '2' , '2']})

print(tabela_2)
```

image.png

**1) Interseção exata entre as tabelas.** (método: how = ‘inner’)

Somente os valores na coluna “Nome” que existem em ambas tabelas aparecem no nosso resultado.

```python
m = pd.merge(tabela_1, tabela_2, how = 'inner', on = 'Nome')
```

```python
print(m)
```

**2) Todas as informações, de ambas tabelas.** (método: how = ‘outer’)

Resultado para os campos que a coluna 'Nome' não tem informação, permanece em branco.

```python
m = pd.merge(tabela_1, tabela_2, how = 'outer')
```

```python
print(m)
```

**3) Demais colunas de uma tabela de acordo com um parâmetro.** (método: how = ‘left’ ou how= ‘right’)

Somente ‘Nomes’ que existam na tabela_1 serão trazidos da tabela_2.
Neste exemplo deixaremos a tabela_1 à esquerda e faremos um merge ‘left’.

```python
m = pd.merge(tabela_2, tabela_1, how = 'left', on = 'Nome')
```

```python
print(m)
```
